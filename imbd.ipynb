```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("users.data",sep="\t",names=["user_id","item_id","rating","timestamp"])
```

```python
data.head()
```

```python
len(data)
```

```python
filmler = pd.read_csv("movie_id_titles.csv")
```

```python
filmler.head()
```

```python
len(filmler)
```

```python
data =pd.merge(data,filmler,on="item_id")
```

```python
data.tail()
```

```python
tablo = data.pivot_table(index="user_id",columns="title",values = "rating")
tablo.head()
```

```python
starwars=tablo["Star Wars (1977)"]
starwars.head()
```

```python
type(starwars)
```

```python
korelasyon = tablo.corrwith(starwars)
```

```python
korelasyon
```

```python
kore=pd.DataFrame(korelasyon,columns=["Correlation"])
kore.dropna(inplace=True)
kore.head()
```

```python
kore.sort_values("Correlation",ascending=False).head(15)
```

```python
ratings=pd.DataFrame(data.groupby("title")["rating"].mean())
ratings.sort_values("rating",ascending=False).head()
```

```python
ratings["oysayisi"]=pd.DataFrame(data.groupby("title")["rating"].count())
```

```python
ratings.head()
```

```python
ratings.sort_values("oysayisi",ascending=False).head()
```

```python
kore.head()
```

```python
kore=kore.join(ratings["oysayisi"])
kore.head()
```

```python
kore[kore["oysayisi"]>100].sort_values("Correlation",ascending=False).head(15)
```
